```python
import pandas as pd
import numpy as np
from keras.preprocessing.image import array_to_img, img_to_array, load_img
import multiprocessing

%matplotlib inline
```

```python
from os import listdir, rmdir, mkdir
from shutil import rmtree
from os.path import isfile, isdir, join

ORI_DIR = '../Facescrub-Faceonly'
CROP_DIR = '../Facescrub-Faceonly-Cropped'
dirs = [d for d in listdir(ORI_DIR) if isdir(join(ORI_DIR, d))]
rmtree(CROP_DIR, ignore_errors=True)
mkdir(CROP_DIR)
```

```python
img_paths = []
```

```python
for d in dirs:
    mkdir(join(CROP_DIR, d))
    files = [join(d, f) for f in listdir(join(ORI_DIR, d)) if isfile(join(ORI_DIR, d, f))]
    img_paths.extend(files)
```

```python
def cropImgAndResize(path):
    try:
        img = load_img(join(ORI_DIR, path))
    except Exception:
        print(path)
        return None
    new_img = img.resize((47,55))
    try:
        new_img.save(join(CROP_DIR, path))
    except Exception:
        print('save', path)
        return None
    
    return new_img.size
```

```python
pool = multiprocessing.Pool(8)
results = pool.map(cropImgAndResize, img_paths)
pool.close()
pool.join()
```
